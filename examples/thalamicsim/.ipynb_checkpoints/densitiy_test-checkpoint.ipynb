```python
import pandas as pd
import numpy as np
import seaborn as sns

import vmcell

import os

import MeMo.compiler.neuron.modules as modules
modules.compile()

c = vmcell.Cell('Test')
c.make()
```

```python
# calculate the number of post-synaptic boutons   
cx  = { 0:(28.9*24.4)/10000.0, 1:(22.5*13)/10000.0, 2:(2.8*1.4)/10000.0 }
rtn = { 0:(3.1*0.8)/10000.0, 1:(2.5*0.8)/10000.0, 2:(2.2*1.4)/10000.0 }
snr = { 0:(0*0.8)/10000.0, 1:(1.1*3.1)/10000.0, 2:(3.9*3.6)/10000.0 }
```

```python
def calculate_nsyn(cell, dens):
    cnt0 = 0
    nsyn = 0
    for sec in cell.icell.basal:
        for seg in sec.allseg():
            idiam = int(seg.diam / 0.5)
            if idiam > 2:
                idiam = 2
            if idiam == 0:
                cnt0 += 1
            nsyn += seg.area() * dens[idiam]
    print(cnt0)
    return nsyn
```

```python
num0 = []
for filemorph in os.listdir('vmcell/morphologies'):
    num0.append(calculate_nsyn( c.mk_cell_model(altmorph=filemorph), snr))
num0 = np.array(num0)
```

```python
print(round(np.mean(num0)))
```

```python
num1 = []
for filemorph in os.listdir('vmcell/morphologies'):
    num1.append(calculate_nsyn( c.mk_cell_model(altmorph=filemorph), rtn))
num1 = np.array(num1)
```

```python
num2 = []
for filemorph in os.listdir('vmcell/morphologies'):
    num2.append(calculate_nsyn( c.mk_cell_model(altmorph=filemorph), cx))
num2 = np.array(num2)
```

```python
print(round(np.mean(num2)))
```
