```python
import numpy as np
import pandas as pd
import MeMo.nwbio as nwbio
import matplotlib.pyplot as plt
```

```python
f = nwbio.FileReader('test_0.nwb')
exec('session_description = ' + f.nwbfile.session_description)
tab = pd.DataFrame(session_description).set_index('key')

for file_key in f.nwbfile.acquisition.keys():
    idx = file_key.split('.')[0]
    tab.loc[idx, 'file_object'] = f
    tab.loc[idx, 'file_key'] = file_key
```

```python
for k1, g in tab.groupby('lesioned_flag'):
    plt.figure(figsize=(15, 7.5*4))
    for i, (k2, r) in enumerate(g.iterrows()):
        plt.subplot(10, 2, i+1)
        plt.title(str(k1) + '|' + str(r['cellid']))
        color = 'red' if r['cellid'] in ([0,1,2] if k1 else [2, 4, 5, 6, 8]) else 'blue'
        plt.plot(*r['file_object'].read(r['file_key']), color=color)
        plt.plot([0, 20000], [-5, -5], color='green')
        plt.ylim([-80, 40])
    plt.tight_layout(pad=2)
    plt.show()
```
