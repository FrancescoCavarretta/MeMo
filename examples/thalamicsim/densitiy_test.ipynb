```python
import pandas as pd
import numpy as np
import seaborn as sns

import vmcell

import os

import MeMo.compiler.neuron.modules as modules
modules.compile()

c = vmcell.Cell('Test')
c.make()
```

```python
def calculate_nsyn(cell, dens):
    cnt0 = 0
    nsyn = 0
    for sec in cell.icell.basal:
        for seg in sec.allseg():
            idiam = int(seg.diam / 0.5)
            if idiam > 2:
                idiam = 2
            if idiam == 0:
                cnt0 += 1
            nsyn += seg.area() * dens[idiam]
    print(cnt0)
    return nsyn
```

```python
def get_density(dens):
    df = pd.DataFrame()
    for _input, _dens in dens.items():
        for _filemorph in os.listdir('vmcell/morphologies'):
            n = calculate_nsyn( c.mk_cell_model(altmorph=_filemorph), _dens)
            i = df.shape[0]
            df.loc[i, 'morphology'] = _filemorph
            df.loc[i, 'input'] = _input
            df.loc[i, 'n syn'] = n
    return df
```

```python
cx  = { 0:24.4/100.0, 1:13/100.0, 2:1.4/100.0 }
rtn = { 0:0.8/100.0, 1:0.8/100.0, 2:1.4/100.0 }
snr = { 0:0.8/100.0, 1:3.1/100.0, 2:3.6/100.0 }

for coll in [cx, rtn, snr]:
    for k in coll.keys():
        coll[k] = coll[k] * coll[k]
        
dens = { 'cx':cx, 'rtn':rtn, 'snr':snr }

d1 = get_density(dens)
```
